## Plot simulated receivers

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def load(file): 
    
    src_dir="/home/andbro/Desktop/CHEESE/SeisSol/output/Northridge/"
    
    cols =  ["Time","xx","yy","zz","xy","yz","xz","u","v","w"]
    
    return  pd.read_csv(src_dir+file,delimiter='  ', header=5, names=cols)

```

```python
def plot1(df): 
    fig, ax = plt.subplots(3,1, figsize=(15,10))

    ax[0].plot(df.u)
    ax[1].plot(df.v)
    ax[2].plot(df.w)
```

```python
def plot2(df):
    fig, ax = plt.subplots(3,1, figsize=(15,10))

    ax[0].plot(df.xx)
    ax[1].plot(df.yy)
    ax[2].plot(df.zz)
```

```python
df = load("mycomputed/test-receiver-00002-00000.dat")

plot1(df)

plot2(df)
```
