## Define Thresholds for light sensors

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# from numpy import nan
# from pathlib import Path
# from obspy import UTCDateTime
```

### Thresholds

Test was done to find light limits/thresholds on 2021-09-02 for WS6

16:45 - 16:46 LT: daylight         --> ~170 lx

16:46 - 16:47 LT: daylight + light --> ~250 lx

16:47 - 16:48 LT: light            --> ~80 lx


```python

path = '/import/freenas-ffb-01-data/romy_archive/2021/BW/WROMY/LX6.D/'

df = pd.read_csv(path+'20210902.txt')

dat = df.iloc[:,2]

### ________________

fig = plt.figure(figsize=(15,5))

plt.plot(dat[700:750].rolling(1).mean(), zorder=2)
plt.scatter(np.arange(700,750), np.array(dat[700:750]), zorder=2)

plt.axhline(20, color="r", zorder=1, ls="-.")
plt.axhline(140, color="r", zorder=1, ls="-.")

plt.grid(ls="--",zorder=1)

plt.show();
```
