```python
def get_data_archive(station,dt,comp):

    path1 = '/bay200/mseed_online/archive/{1}/BW/{0}/EH{2}.D/BW.{0}..EH{2}.D.{1}.{3}'.format(station, dt.year, comp, dt.julday) 
    path2 = '/bay200/mseed_online/archive/{1}/BW/{0}/LOG.L/BW.{0}..LOG.L.{1}.{3}'.format(station, dt.year, comp, dt.julday) 
    
    return obs.read(path1)

#data = get_data_archive('MSBB',dt,'Z')

#data.plot(starttime=dt,endtime=dt+delta_T);

#data.plot(type='dayplot');
```

```python
from andbro__filter_traces import __filter_traces


dt=obs.UTCDateTime('2020-08-18 10:00'); delta_T=40*60

data = get_data_archive('MSBB',dt,'Z')

#data.remove_response(inventory=inv)

data1 = __filter_traces(data,[0.1,80])
#data1 = data.copy()

data2 = data1.copy()
data2.trim(starttime=dt+0.2*60,endtime=dt+0.8*60)

data3 = data1.copy()
data3.trim(starttime=dt+5.4*60,endtime=dt+6*60)



data1.plot(starttime=dt,endtime=dt+10*60);
#data1[0].spectrogram()

data2.plot();
data2[0].spectrogram(log=True);

data3.plot();
data3[0].spectrogram(log=True);


```
