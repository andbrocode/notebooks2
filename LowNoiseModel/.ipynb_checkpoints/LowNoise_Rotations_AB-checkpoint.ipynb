## Low Noise Model for rotations (and strain)


![](ppsd.png)


* Extend the existing **Low/High Noise models (LNM/HNM)** for ground  acceleration to **rotation (rate) and strain(?)**

* Under the assumption of primarily **fundamental mode Love and Rayleigh waves** in the noise convert LNM/HNM from acceleration to rotation (rate) and strain
* For each frequency develop **frequency (likelihood) of phase velocities** from as many 1D velocity models as possible 
* Final goal: „fuzzy“ LNM/HNM for rotation (and strain) based on educated choice of phase velocities
* Support by **observations** wherever possible (e.g. G-ring, ROMY, array-derived rotation and strain, e.g., PFO)
* Optional: Extend to **Mars, Moon**


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 

from obspy import read
from obspy.signal import PPSD
from obspy.imaging.cm import pqlx
from obspy.io.xseed import Parser
```

```python
# Load low noise model and plot it
NOISE_MODEL_FILE = "data/noise_models.npz"

data = np.load(NOISE_MODEL_FILE)
periods = data['model_periods']
nlnm = data['low_noise']
nhnm = data['high_noise']


plt.figure(figsize=(15,5))

plt.semilogx(periods,nlnm,periods,nhnm)
plt.xlabel('Period')
plt.ylabel('Spectral density (in dB)')
plt.grid(ls=":");
plt.xlim(periods[-1],periods[0])
plt.show();
```

### Convert to acceleration

The noise model is given in dB, where 

$
P[dB] = 10 \log_{10}[P_{acc}/1(m/s^2)^2/Hz] \ .
$

We obtain the power spectral density in $[(m/s^2)^2/Hz]$ by

$
P_{acc} =  10^{P[dB]/10} \ m^2/s^4/Hz
$

and the square-root psd as 

$
P^{sqrt}_{acc} =  \sqrt{10^{P[dB]/10}} \ m/s^2/\sqrt{Hz}
$



```python
# Convert to acceleration
snlnm = np.sqrt(10**(nlnm/10))
snhnm = np.sqrt(10**(nhnm/10))


plt.figure(figsize=(15,5))

font = 14

plt.loglog(periods, snlnm, label="Low Noise Model")
plt.loglog(periods, snhnm, label="High Noise Model")


plt.grid(True, which="both", ls="-", alpha=0.5)
plt.xlabel('Period', fontsize=font)
plt.ylabel(r'ASD  ($\frac{m}{s^2 \sqrt{Hz}}$)', fontsize=font)
plt.xlim((0.1, 1000))
plt.ylim((0., 2e-4))
plt.legend()
plt.show();
```

### Convert to rotation rate

We assume a constant surface wave phase velocity of $c = 3000 m/s$ and assume Rayleigh waves, then the rotation rate spectral density $P_\omega$ is obtained by (ideally we use a dispersion relation c(T))

$
P_{omega} = P^s_{acc}/c(T)
$

### Converting acceleration rPSD with Rayleigh wave dispersion curve

```python
# Approximate dispersion curve (from Shearer and other sources)
T = np.array([0.1,10,25,50,100,150,200,250,300,500])
cR = np.array([300.,3500.,3800.,4000.,4150.,4250.,4500.,4750,5025.,7000.])
uncertainties = np.array([5,5,5,5,10,10,10,20,20,25])


# reverse periods array for interpolation
periods = periods[::-1]
snlnm = snlnm[::-1]
snhnm = snhnm[::-1]


# Limit to T < 500, rename 
p = periods[:616]
lnm = snlnm[:616]
hnm = snhnm[:616]


# Interpolate dispersion curve
cRi = np.interp(p,T,cR)
devi = np.interp(p,T,uncertainties)
# devi = np.ones(len(cRi)) * 20 # in % 
```

```python
def __convert_to_rotation(periods, alnm, ahnm, phase_velocities, uncertainties=None):
    
    df = pd.DataFrame()
    
    df['periods'] = periods
    
    df['rlnm'] = alnm/phase_velocities**2/4
    df['rhnm'] = ahnm/phase_velocities**2/4

    df['cR'] = phase_velocities
    df['dev'] = uncertainties
    
    if uncertainties is not None:
        df['rlnm_conf+'] = alnm/(phase_velocities*(1+uncertainties/100))**2/4
        df['rlnm_conf-'] = alnm/(phase_velocities*(1-uncertainties/100))**2/4

        df['rhnm_conf+'] = ahnm/(phase_velocities*(1+uncertainties/100))**2/4
        df['rhnm_conf-'] = ahnm/(phase_velocities*(1-uncertainties/100))**2/4    
    
    return df

df = __convert_to_rotation(p, lnm, hnm, cRi, uncertainties=devi)
```

```python
def __makeplot(df):
    
    font = 14
    
    plt.figure(figsize=(15,5))
    plt.plot(df['periods'], df['cR'])
    plt.fill_between(df['periods'], df['cR']*(1-df['dev']/100), df['cR']*(1+df['dev']/100), zorder=1, alpha=0.5, color='grey')
    plt.xlabel('Period', fontsize=font)
    plt.ylabel('Phase velocity (m/s)', fontsize=font)
    plt.xlim((min(df['periods']), 500))
    plt.title('Rayleigh wave dispersion', fontsize=font)
    plt.show();

    plt.figure(figsize=(15,5))
    plt.plot(df['periods'], df['rlnm'], zorder=2)
    plt.fill_between(df['periods'], df['rlnm_conf-'], df['rlnm_conf+'], zorder=1, alpha=0.5, color='grey')
    plt.yscale('log')
    plt.xscale('log')
    plt.xlabel('Period', fontsize=font)
    plt.ylabel(r'rASD ($\frac{rad}{s \sqrt{Hz}}$)', fontsize=font)
    plt.xlim((min(df['periods']), 500))
    plt.title('Estimated rotational low noise model with confidence interval', fontsize=font)
    plt.show();


    plt.figure(figsize=(15,5))
    plt.plot(df['periods'], df['rhnm'], zorder=2)
    plt.fill_between(df['periods'], df['rhnm_conf-'], df['rhnm_conf+'], zorder=1, alpha=0.5, color='grey')
    plt.yscale('log')
    plt.xscale('log')
    plt.xlabel('Period', fontsize=font)
    plt.ylabel(r'rASD ($\frac{rad}{s \sqrt{Hz}}$)', fontsize=font)
    plt.xlim((min(df['periods']), 500))
    plt.title('Estimated rotational high noise model with confidence interval', fontsize=font)
    plt.show();
    
__makeplot(df)
```

```python
def __conversion(data, informat, outformat):
    '''
    data:                  array 
    informat/outformat:    ASD, PSD, DB
    '''
    from numpy import log10
    
    
    if informat == 'ASD' and outformat == 'PSD':
        return data**2

    elif informat == 'ASD' and outformat == 'DB':
        return 20 * log10(data)
    
    elif informat == 'PSD' and outformat == 'DB':
        return 10 * log10(data)
    
    elif informat == 'DB' and outformat == 'PSD':
        return 10**(data/10)
    
    elif informat == 'DB' and outformat == 'ASD':
        return 10**(data/20)  
```

```python
df_db = df.copy()

df_db['rlnm'] = __conversion(df_db['rlnm'], 'ASD', 'DB')
df_db['rhnm'] = __conversion(df_db['rhnm'], 'ASD', 'DB')

df_db['rlnm_conf+'] = __conversion(df_db['rlnm_conf+'], 'ASD', 'DB')
df_db['rlnm_conf-'] = __conversion(df_db['rlnm_conf-'], 'ASD', 'DB')

df_db['rhnm_conf+'] = __conversion(df_db['rhnm_conf+'], 'ASD', 'DB')
df_db['rhnm_conf-'] = __conversion(df_db['rhnm_conf-'], 'ASD', 'DB')
```

```python
# Convert rPSD with dispersion curve
def __makeplot_joint(df):

    plt.figure(figsize=(15,5))

    plt.plot(df['periods'], df['rlnm'], zorder=2, label="rLNM")
    plt.fill_between(df['periods'], df['rlnm_conf-'], df['rlnm_conf+'], zorder=1, alpha=0.5, color='grey')

    plt.plot(df['periods'], df['rhnm'], zorder=2, label="rHNM")
    plt.fill_between(df['periods'], df['rhnm_conf-'], df['rhnm_conf+'], zorder=1, alpha=0.5, color='grey')
    
    plt.grid(True, which="both", ls="-", alpha=0.5)
    plt.yscale('log')
    plt.xscale('log')
    plt.xlabel('Period', fontsize=font)
    plt.ylabel(r'rASD ($\frac{rad}{s \sqrt{Hz}}$)', fontsize=font)
    plt.xlim((min(df['periods']), 500))
    plt.title('Estimated Rotational Low/Nigh Noise Model with Confidence Interval', fontsize=font)
    plt.legend()
    plt.show();
    
__makeplot_joint(df)
```

```python
def __makeplot_joint_DB(df):

    plt.figure(figsize=(15,5))

    plt.plot(df['periods'], df['rlnm'], zorder=2, label="rLNM")
    plt.fill_between(df['periods'], df['rlnm_conf-'], df['rlnm_conf+'], zorder=1, alpha=0.5, color='grey')

    plt.plot(df['periods'], df['rhnm'], zorder=2, label="rHNM")
    plt.fill_between(df['periods'], df['rhnm_conf-'], df['rhnm_conf+'], zorder=1, alpha=0.5, color='grey')
    
    plt.grid(True, which="both", ls="-", alpha=0.5)
#     plt.yscale('log')
    plt.xscale('log')
    plt.xlabel('Period', fontsize=font)
    plt.ylabel(r'rPSD (dB  rel. to 1 $\frac{rad^2}{s^2 Hz}$)', fontsize=font)
    plt.xlim((min(df['periods']), 500))
    plt.title('Estimated Rotational Low/Nigh Noise Model with Confidence Interval', fontsize=font)
    plt.legend()
    plt.show();

__makeplot_joint_DB(df_db)
```
