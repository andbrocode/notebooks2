## Low Noise Model for rotations (and strain)


![](ppsd.png)


* Extend the existing **Low/High Noise models (LNM/HNM)** for ground  acceleration to **rotation (rate) and strain(?)**

* Under the assumption of primarily **fundamental mode Love and Rayleigh waves** in the noise convert LNM/HNM from acceleration to rotation (rate) and strain
* For each frequency develop **frequency (likelihood) of phase velocities** from as many 1D velocity models as possible 
* Final goal: „fuzzy“ LNM/HNM for rotation (and strain) based on educated choice of phase velocities
* Support by **observations** wherever possible (e.g. G-ring, ROMY, array-derived rotation and strain, e.g., PFO)
* Optional: Extend to **Mars, Moon**


```python
# Import lib
import numpy as np
from obspy import read
from obspy.signal import PPSD
from obspy.imaging.cm import pqlx
from obspy.io.xseed import Parser
import matplotlib.pyplot as plt
```

```python
# Load low noise model and plot it
NOISE_MODEL_FILE = "data/noise_models.npz"

data = np.load(NOISE_MODEL_FILE)
periods = data['model_periods']
nlnm = data['low_noise']
nhnm = data['high_noise']

plt.semilogx(periods,nlnm,periods,nhnm)
plt.xlabel('Period')
plt.ylabel('Spectral density (in dB)')
plt.show()
```

### Convert to acceleration

The noise model is given in dB, where 

$
P[dB] = 10 \log_{10}[P_{acc}/1(m/s^2)^2/Hz] \ .
$

We obtain the power spectral density in $[(m/s^2)^2/Hz]$ by

$
P_{acc} =  10^{P[dB]/10} \ m^2/s^4/Hz
$

and the square-root psd as 

$
P^{sqrt}_{acc} =  \sqrt{10^{P[dB]/10}} \ m/s^2/\sqrt{Hz}
$



```python
# Convert to acceleration
snlnm = np.sqrt(10**(nlnm/10))
snhnm = np.sqrt(10**(nhnm/10))


plt.semilogx(periods,snlnm)
plt.xlabel('Period')
plt.ylabel('PSD low noise(in m/s^2/sqrt(Hz))')
plt.xlim((0.1, 1000))
plt.ylim((0., 1e-7))
plt.title(' Low Noise model ')
plt.show()

plt.semilogx(periods,snhnm)
plt.xlabel('Period')
plt.ylabel('PSD high noise (in m/s^2/sqrt(Hz))')
plt.xlim((0.1, 1000))
plt.ylim((0., 2e-5))
plt.title(' High Noise model ')
plt.show()
```

### Convert to rotation rate

We assume a constant surface wave phase velocity of $c = 3000 m/s$ and assume Rayleigh waves, then the rotation rate spectral density $P_\omega$ is obtained by (ideally we use a dispersion relation c(T))

$
P_{omega} = P^s_{acc}/c(T)
$

### Converting acceleration rPSD with Rayleigh wave dispersion curve

```python
# Approximate dispersion curve (from Shearer and other sources)
T = [0.1, 10,25,50,100,150,200,250,300,500]
cR = [300., 3500., 3800.,4000., 4150.,4250., 4500.,4750,5025., 7000.]

# reverse periods array for interpolation
periods = periods[::-1]
snlnm = snlnm[::-1]
snhnm = snhnm[::-1]


# Limit to T < 500, rename 
p = periods[:616]
lnm = snlnm[:616]
hnm = snhnm[:616]

# Interpolate dispersion curve
cRi = np.interp(p,T,cR)

# Test interpolated  dispersion
plt.plot(p,cRi)
plt.xlabel('Period')
plt.ylabel('Phase velocity (m/s)')
plt.xlim((0.1, 500))
plt.title(' Rayleigh wave dispersion ')
plt.show();

# Convert rPSD with dispersion curve
plt.loglog(p,lnm/cRi,p,lnm/(cRi*0.8),p,lnm/(cRi*1.2))
plt.xlabel('Period')
plt.ylabel('rPSD low noise(in rad/s/sqrt(Hz))')
plt.xlim((0.1, 500))
plt.title('Provisional rotational low noise model (+/-20%)')
#plt.ylim((0., 0.4e-10))
plt.show();

plt.loglog(p,hnm/cRi,p,hnm/(cRi*0.8),p,hnm/(cRi*1.2))
plt.xlabel('Period')
plt.ylabel('rPSD high noise (in rad/s/$sqrt(Hz)$)')
plt.xlim((0.1, 500))
plt.title('Provisional rotational high noise model(+/-20%)')
#plt.ylim((0., 1e-8))
plt.show();
```

```python
# Single plot

# Convert rPSD with dispersion curve
plt.loglog(p,lnm/cRi,p,lnm/(cRi*0.8),p,lnm/(cRi*1.2),p,hnm/cRi,p,hnm/(cRi*0.8),p,hnm/(cRi*1.2))
plt.xlabel('Period')
plt.ylabel('rPSD low noise(in rad/s/sqrt(Hz))')
plt.xlim((0.1, 500))
plt.title('Provisional rotational low/high noise model (+/-20%)')
#plt.ylim((0., 0.4e-10))
plt.savefig('lhnmodel',dpi=300)
plt.show()
```
