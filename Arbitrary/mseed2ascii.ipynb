## write mseed to ascii

```python
from obspy.core import read

path = '/home/andbro/Documents/ROMY/monobeams/monobeam_data/'
name = 'DROMY_FJV_2021-02-05.mseed'

stream = read(path+name+'.mseed')

stream.write(path+name+'.ascii', format='SLIST')

# subprocess.run(["sed" "-i" "'s/\t/,/g'" name], shell=True)

```

```python
from pandas import read_csv

df = read_csv(path+'DROMY_F2V_2021-02-05.ascii', delimiter=",", header=1, names=['a','b','c','d','e','f'])
df

df.a.plot()
```
