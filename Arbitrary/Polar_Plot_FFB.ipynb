```python
%matplotlib inline


from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt


lon_OBS=11.277
lat_OBS=48.165

lon_G=12.8780
lat_G=49.1444
```

```python
#%matplotlib qt5

fig, ax = plt.subplots(1, 1, figsize = (15, 10))

wide, long = [24e6,20e6] 

map1 = Basemap(projection='aeqd',lat_0=lat_OBS,lon_0=lon_OBS,fix_aspect=True,width=wide,height=long,\
               area_thresh=1000,resolution='l')


### read a shapefile 
#map.readshapefile('/home/brotzer/resources/NE_countries/ne_10m_admin_0_countries', name='world', drawbounds=False, color='gray')


''' draw second map for grid '''
map2 = Basemap(projection='aeqd',lat_0=90,lon_0=180,fix_aspect=True,width=wide,height=long,area_thresh=1000)
map2.drawparallels(range(-90, 90, 30),dashes=[1,0], fontsize=8,color='grey')
map2.drawmeridians(range(0, 360, 30),labels=[1,1,1,1],dashes=[1,0],fontsize=15,color='grey',labelstyle='+/-')

''' draw map items '''
map1.drawcoastlines(linewidth=0.5,antialiased=1)
map1.drawmapboundary(fill_color='lightblue')
map1.fillcontinents(color='lightgrey',lake_color='lightblue')
map1.drawcountries()

''' locate and annotate ROMY ring-laser '''
x_obs, y_obs = map1(lon_OBS,lat_OBS)
map1.plot(x_obs,y_obs,marker='o',color='r')
#plt.annotate('ROMY',xy=(map1(lon_OBS,lat_OBS)),xytext=(map1(15,50)),color='r',arrowprops=dict(arrowstyle="->",color='r'))

plt.title('Equidistant-Azimutal Plot for ROMY ',fontsize=17,pad=40)

plt.show();
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize = (10, 10))


map = Basemap(llcrnrlon=-12,llcrnrlat=40,urcrnrlon=40.,urcrnrlat=55.,
             resolution='l', projection='cass', lat_0 = 39.5, lon_0 = 0.)


'''add topography data''' 
#map.bluemarble()
map.etopo()


''' draw map items '''
map.drawcoastlines()
map.drawcountries(linewidth=1.0, linestyle='solid', color='k', antialiased=1, ax=None, zorder=None)

map.drawmapscale(-7., 45.8, 11., 48., 500, barstyle='fancy',units='km')


''' locate and annotate ROMY ring-laser '''
x,y = map(lon_OBS,lat_OBS)
map.plot(x,y,marker='^',color='k',markersize=3)
plt.annotate('ROMY',xy=(map(lon_OBS,lat_OBS)),xytext=(map(7,50)),color='r',fontsize=15,arrowprops=dict(arrowstyle="->",color='r'))


''' locate and annotate G ring-laser '''
x,y = map(lon_G,lat_G)
map.plot(x,y,marker='^',color='k',markersize=3)
plt.annotate('G-Ring',xy=(map(lon_G,lat_G)),xytext=(map(15,52)),color='b',fontsize=15,arrowprops=dict(arrowstyle="->",color='b'))


''' draw meridians and parallels '''
map.drawmeridians(range(0, 360, 10),dashes=[1,4],labels=[0,0,1,1], fontsize=15,color='lightgrey')
map.drawparallels(range(0, 360, 10),dashes=[1,4],labels=[1,1,0,0], fontsize=15,color='lightgrey')

plt.show();
```
