''' Chandler Wobble '''

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
timeline=16 # years
T1=365 # days
T2=433 # days (Wikipedia)

time=np.arange(0,365*timeline,1)

c=4*np.sin(2*np.pi/T1*time)+np.sin(2*np.pi/T2*time)


fig, ax = plt.subplots()

ax.plot(time,c)

ax.set_xlabel('years')
ax.set_ylabel('amplitude')

ax.set_xticks(np.arange(0,timeline*365+365,365))
ax.set_xticklabels(["$%i$" % y for y in np.arange(0,timeline+1,1) ]);

for index, label in enumerate(ax.xaxis.get_ticklabels()):
    if index % 2 != 0:
        label.set_visible(False)
```
