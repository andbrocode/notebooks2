# Array Response Function for different Configuration & Frequency

### Importing

```python

import numpy as np
import matplotlib.pyplot as plt

from obspy.clients.fdsn import Client 
from obspy import UTCDateTime
from obspy.imaging.cm import obspy_sequential
from obspy.signal.array_analysis import array_transff_wavenumber
```

### Setting Variables

```python
## set input path
ipath = "/home/andbro/notebooks/backprojections/"

# set limits for wavenumber differences to analyze, corresponding to frequency variations
klim = 20.
dx = 100.

kxmin = -klim
kxmax = klim
kymin = -klim
kymax = klim
kstep = klim / dx
```

### Defining Methods

```python

def __makeplot_array(array, transff):
    
    array_center = [np.mean(array[:,0]), np.mean(array[:,1])]
    
    fig, axes = plt.subplots(1,2, figsize=(14,6))

    ## ________________________________________________
    axes[0].plot(array[:,0],array[:,1],'kv')
    axes[0].scatter(array_center[0], array_center[1], color='red', marker='o')

    axes[0].set_xlabel('Longitude')
    axes[0].set_ylabel('Latitude')
    axes[0].set_title('Square Array configuration')

    
    
    ## ________________________________________________
    ax2 = axes[1].pcolor(np.arange(kxmin, kxmax + kstep * 1.1, kstep) - kstep / 2.,
               np.arange(kymin, kymax + kstep * 1.1, kstep) - kstep / 2.,
               transff.T, cmap=obspy_sequential)
    
    plt.colorbar(ax2)
    ax2.set_clim(0., 1.)
    axes[1].set_xlim(kxmin, kxmax)
    axes[1].set_ylim(kymin, kymax)
    axes[1].set_title('ARF Square Array')
    
    plt.show()    
    
```

### Plotting Array Geometry and Response

```python
##load the Oklahoma array to monitor induced seismicity
array_ok = np.loadtxt(f"{ipath}array_ok.txt")


## compute transfer function as a function of wavenumber difference
transff = array_transff_wavenumber(array_ok, 
                                   klim, 
                                   kstep, 
                                   coordsys='lonlat'
                                  )

    
__makeplot_array(array_ok, transff)
```

```python
##cross linear array
array_linear_cross = np.loadtxt(f"{ipath}array_linear_cross.txt")

##change the size of the array, units in km
array_linear_cross /= 2e4 


## compute transfer function as a function of wavenumber difference
transff = array_transff_wavenumber(array_linear_cross, 
                                   klim, 
                                   kstep, 
                                   coordsys='lonlat'
                                  )

__makeplot_array(array_linear_cross, transff)
```

```python
##GA array in southern CA
c = Client("IRIS")

tbeg = UTCDateTime("2011-06-06T12:16:00.000")
tend = UTCDateTime("2011-06-06T12:18:00.000")


def __get_array_stations(seed_id, tbeg, tend):

    net, sta, loc, cha  = seed_id.split(".")

    inv_check = c.get_stations(network=net, 
                               station=sta,
                               channel=cha,
                               starttime=tbeg,
                               endtime=tend
                              )
    #inv_check.plot() 
    inv_check.plot(projection="local");
    
    codes, lon, lat, elev = [], [], [], []
    for network in inv_check:
        for i, station in enumerate(network):
            ##station_coordinates.append((network.code, station.code, station.latitude, station.longitude, station.elevation))
#             stations.append((station.code,
#                              station.longitude,
#                              station.latitude))
            codes.append(station.code)
            lon.append(station.longitude)
            lat.append(station.latitude)
            elev.append(station.elevation)

    
#    return np.column_stack((code,lon,lat))
#    return np.concatenate((code,lon,lat),axis=1)
    return codes, np.array([lon, lat, elev])


array_stations, array_coords = __get_array_stations("ZZ.GA*..SHZ", tbeg, tend)

```

```python

##convert elevation to km
array_coords[2][:] = array_coords[2][:]/1000 


# compute transfer function as a function of wavenumber difference
transff = array_transff_wavenumber(array_coords, klim, kstep, coordsys='lonlat')


__makeplot_array(array_coords, transff)
```
