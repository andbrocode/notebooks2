```python
import pandas as pd
import numpy as np
```

```python
path = "/home/andbro/Documents/LMU_Lehre/Einführung_in_die_Geophysik_2021/modify/"

basename = input("Enter basename of file: ")

# data = pd.read_csv(path+basename+".csv", delimiter=",", names=['nachname','vorname','mail','kurs','aufgabe'])

data1 = pd.read_csv(path+f"{basename}A.csv", delimiter=",", names=['nachname','vorname','mail','kurs','aufgabe'])
data2 = pd.read_csv(path+f"{basename}B.csv", delimiter=",", names=['nachname','vorname','mail','kurs','aufgabe'])

data = pd.concat([data1, data2])

del data['mail'] 
data.drop(0, inplace=True)

# if {basename[-1]} == 'B':
#     data = data[data.kurs == f'Kurs B (Donnerstag)']
# elif {basename[-1]} == 'A':
#     data = data[data.kurs == f'Kurs A (Dienstag)']

del data['kurs']

data = data[data.aufgabe != "Nicht abgestimmt"].reset_index()

data
```

```python
aufgaben = list(data.aufgabe.unique())

# names = pd.read_csv(path+f"names{basename[-1].lower()}.csv", names=["nachname", "vorname"])

names = pd.read_csv(path+f"names.csv", names=["nachname", "vorname"])
```

```python
done = np.zeros(len(names))

for aufgabe in aufgaben:
    for i in range(len(names)):
        sub = data.loc[(data['vorname'] == names.vorname[i]) & (data['nachname'] == names.nachname[i])]
        if sub.aufgabe.str.match(str(aufgabe)).any():
            done[i] = 1
    names[aufgabe] = done
    names[aufgabe].apply(int)
    done = np.zeros(len(names))
names
```

```python
# names.to_excel(path+f"{basename}_done.xlsx", index=False)
names.to_csv(path+f"{basename}_done.csv", index=False)
```
