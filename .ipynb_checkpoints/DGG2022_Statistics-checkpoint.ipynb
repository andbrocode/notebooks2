```python
themes = submission1[pd.notnull(submission1['Abstracteinreichungen sortiert nach Themen'])]
# themes = themes[~themes.iloc[:,1].isin(['S1', 'S2', 'S3', 'S4'])]
themes = themes[~themes.iloc[:,2].isin(["Schwerpunktthemen", "Sessions"])]

themes
```

```python
def __makeplot_bar(config):
    
    '''
    Config: 
        - xdata         data on x-axis
        - ydata         data on y-axis
        - xth           nth xtick that is shown
        - ylabel        label for y-axis
        - title         title of plot
        
    '''
    
    import matplotlib.pyplot as plt    

    # Creating histogram
    fig, ax = plt.subplots(1, 1, figsize =(12, 5), tight_layout = True)

    font = 13
    NN = len(config['xdata'])

    # Remove axes splines
    for s in ['top', 'bottom', 'left', 'right']:
        ax.spines[s].set_visible(False)

    # Remove x, y ticks
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')

    # Add padding between axes and labels
    ax.xaxis.set_tick_params(pad = 5)
    ax.yaxis.set_tick_params(pad = 10)

    # Add x, y gridlines
#     ax.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.6, zorder=0)
    ax.grid(axis='y', color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.6, zorder=0)


    # Creating histogram
    bars = range(0, NN)
    ax.bar(bars, config['ydata'], width=0.8, align='center', alpha=0.8, label=str(config['xdata']))


    # Set ticks
    datelabel = config['xdata'][::config['xth']]
    dateticks = bars[::config['xth']]

    ax.set_xticklabels(datelabel, fontsize=font)
    ax.set_xticks(dateticks)

    
    # Adding extra features   
    plt.ylabel(config['ylabel'], labelpad=5, fontsize=font)
    plt.title(config['title'], pad=10, fontsize=font+1)
          
    plt.figtext(0.1, 0.83, f"Gesamt: {sum(config['ydata']).astype(int)}", fontsize=font-1)
    # Show plot
    plt.show();

    return fig
```

```python
config = {}

config['ydata'] = np.array(themes['Angenommen'].astype(float))
config['xdata'] = np.array(themes['ThemenNr.'])
config['xth']   = 1        
config['ylabel'] = "Anzahl" 
config['title'] = "DGG2022 - Angenommene Beiträge nach Themenschwerpunkt"        

__makeplot_bar(config);
```

```python
def __makeplot_bar2(config):
    
    '''
    Config: 
        - xdata         data on x-axis
        - ydata         data on y-axis
        - xth           nth xtick that is shown
        - ylabel        label for y-axis
        - title         title of plot
        
    '''
    
    def autolabel(rects):
        """
        Attach a text label above each bar displaying its height
        """
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2., .90*height,
                    '%d' % int(height),
                    ha='center', va='bottom', color="white")

    
    import matplotlib.pyplot as plt    

    # Creating histogram
    fig, ax = plt.subplots(1, 1, figsize =(5, 5), tight_layout = True)

    font = 13
    NN = len(config['xdata'])

    # Remove axes splines
    for s in ['top', 'bottom', 'left', 'right']:
        ax.spines[s].set_visible(False)

    # Remove x, y ticks
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')

    # Add padding between axes and labels
    ax.xaxis.set_tick_params(pad = 5)
    ax.yaxis.set_tick_params(pad = 10)

    # Add x, y gridlines
#     ax.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.6, zorder=0)
    ax.grid(axis='y', color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.6, zorder=0)


    # Creating histogram
    bars = range(0, NN)
    bars1 = ax.bar(bars, config['ydata'], width=0.7, align='center', alpha=0.8)


    # Set ticks
    datelabel = config['xdata'][::config['xth']]
    dateticks = bars[::config['xth']]

    ax.set_xticklabels(datelabel, fontsize=font)
    ax.set_xticks(dateticks)

#     ax.bar_label(ax.containers[0])
    autolabel(bars1)
    
    # Adding extra features   
    plt.ylabel(config['ylabel'], labelpad=5, fontsize=font)
    plt.title(config['title'], pad=10, fontsize=font+1)
          
    plt.figtext(0.5, 0.73, f"Gesamt: {sum(config['ydata']).astype(int)}", fontsize=font-1)
    # Show plot
    plt.show();

    return fig
```

```python
config = {}

config['ydata'] = np.array(gender['Anzahl der Einreichungen'].astype(float))
config['xdata'] = np.array(gender['Anrede'])
config['xth']   = 1        
config['ylabel'] = "Anzahl" 
config['title'] = "DGG2022 - \n Abstract Einreichungen nach Geschlecht\n"        

__makeplot_bar2(config);
```

```python
def __makeplot_bar3(config):
    
    '''
    Config: 
        - xdata         data on x-axis
        - ydata         data on y-axis
        - xth           nth xtick that is shown
        - ylabel        label for y-axis
        - title         title of plot
        
    '''
    
    import matplotlib.pyplot as plt    

    # Creating histogram
    fig, ax = plt.subplots(1, 1, figsize =(12, 5), tight_layout = True)

    font = 13
    NN = len(config['xdata'])

    # Remove axes splines
    for s in ['top', 'bottom', 'left', 'right']:
        ax.spines[s].set_visible(False)

    # Remove x, y ticks
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')

    # Add padding between axes and labels
    ax.xaxis.set_tick_params(pad = 5)
    ax.yaxis.set_tick_params(pad = 10)

    # Add x, y gridlines
#     ax.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.6, zorder=0)
    ax.grid(axis='y', color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.6, zorder=0)


    # Creating histogram
    bars = range(0, NN)
    ax.bar(bars, config['ydata'],  width=0.8, align='center', alpha=0.9)


    # Set ticks
    datelabel = config['xdata'][::config['xth']]
    dateticks = bars[::config['xth']]

    ax.set_xticklabels(datelabel, fontsize=font, rotation=60, ha="right")
    ax.set_xticks(dateticks)


    # Adding extra features   
    plt.ylabel(config['ylabel'], labelpad=5, fontsize=font)
    plt.title(config['title'], pad=10, fontsize=font+1)
          
    plt.figtext(0.1, 0.83, f"Gesamt: {sum(config['ydata']).astype(int)}", fontsize=font-1)
    # Show plot
    plt.show();

    return fig
```

```python
config = {}

config['ydata'] = np.array(subdate['Anzahl der Einreichungen'].astype(float))
config['xdata'] = np.array(subdate['Datum'])
config['xth']   = 1        
config['ylabel'] = "Anzahl" 
config['title'] = "DGG2022 - Abstract Einreichungen nach Geschlecht"        

__makeplot_bar3(config);
```
