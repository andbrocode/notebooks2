## Create Synthetics

## Load Libraries & Modules

```python
import matplotlib.pyplot as plt

from submodules.CreateSyntheticEventV4 import __create_synthetic_event_v4
from submodules.MakeplotTraceAndSpectrum import __makeplot_trace_and_spectrum
from submodules.CreateLinearChirp import __create_linear_chirp
from submodules.Tapering import __tapering
```

## Configurations

```python
config = {}

config['T'] = 1600 ## seconds
config['sps'] = 5000 ## samples per second
config['f_lower'], config['f_upper'] = 0.01, 0.1 ## in Hz
```

## Create a Synthetic Trace

create a synthetic trace / event based on a convolution of random spikes and either a 'chirp' or 'ricker' wavelet. 

```python
## create trace
event, event_time =__create_synthetic_event_v4(config, 
                                               signal_type='chirp', 
                                               noise_level=None, 
                                               set_taper=True, 
                                               set_normalize=True,
                                              )


## plotting
__makeplot_trace_and_spectrum(event, event_time, fmax=2*config['f_upper']);
```

## Create a Synthetic Chirp

```python
## generate chirp signal
chirp, timeline = __create_linear_chirp(config['T'], config['sps'], config['f_lower'], config['f_upper'])

## taper chirp to avoid Gibb's effect
chirp = __tapering(chirp, taper_type='hann', percent=0.2)

## plotting
__makeplot_trace_and_spectrum(chirp, timeline, fmax=2*config['f_upper'] );
```

## Storing Trace

```python
## select array to store: 'chirp' or 'event'
synthetic_signal = event
```

```python
import pickle 

config['opath'] = "/home/andbro/notebooks/Sagnac_processing/data/"

## double check 
store = input("Really store traces (y/n)?") or False

if store.lower() in ['y', 'yes']:
    
    config['oname'] = f"synthetic_T{config['T']}_fs{config['sps']}_f{config['f_lower']}-{config['f_upper']}"
    
    ofile = open(f"{config['opath']}{config['oname']}.syn","wb")
    pickle.dump(synthetic_signal, ofile)
    ofile.close()
    
    config['oname'] = ""
    
    ofile = open(f"{config['opath']}{config['oname']}.conf","wb")
    pickle.dump(config, ofile)
    ofile.close()    
```
