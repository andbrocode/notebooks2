```python
# Imports
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import matplotlib.pylab as plt

from scipy import interpolate, signal
from scipy.signal import hilbert, chirp
from scipy import signal
#from fitsin import fit_sin
import obspy

from andbro__querrySeismoData import __querrySeismoData
from obspy.clients.fdsn import RoutingClient, Client

%matplotlib inline
```

### Ring Laser statics

```python
# Test Rotational Signals for Demodulation: Determine Sagnac Carrier Frequency

# Sagnac Frequency G ring
lat = 49.16   # deg
A = 16.0 # square meters
P = 16.0 # meters 
l = 632.8 * 1e-9 
d = 23*3600+56*60+4.1  # length of day
O = 2*np.pi/d # Earth's rotation rates

# Calculate Sagnac frequency
fSagnac = 4*A/(l*P)*O*np.sin(lat/180.*np.pi)

print('Sagnac frequency (G-ring) : ',fSagnac,' Hz')
```

### Synthesizing the Sagnac Signal

* Interpolate rotation seismogram to final sampling rate
* Generate frequency modulation signal (to be demodulated)
https://en.wikipedia.org/wiki/Frequency_modulation
* Good derivation is here: https://en.wikibooks.org/wiki/Communication_Systems/Frequency_Modulation
* Hard to listen to but ends up with the equation I use: https://www.youtube.com/watch?v=If1tezEvZsk

Equation used to generate the FM signal s(t) :

$
s(t) = A \cos( f_{Sagnac} +  2 \pi S_f \int_0^t f_{mod}(t) dt)
$

here $S_f$ is the frequency sensitivity (I don t think we need this, it is set to 1), $f_{mod}$ is the modulated signal (the encoded information). 

In the case of a rotation rate signal this corresponds to

$
f(t) = f_{Sagnac} + f_{mod}(t)
=
f(t) = f_{Sagnac} + \frac{4A}{\lambda P} \ \Omega_s (t)
$

where $\Omega_s (t)$ is the rotation rate signal. 

```python
######################################
# Generate rotation perturbation signals

# Change case to alter forward signal
icase = 5

# Case 0: Original chirp: frequency increases from 20 Hz to 100 Hz and apply an amplitude modulation.
if icase == 0:
    duration = 1.0
    fs = 400.0
    nt = int(fs*duration)
    t = np.arange(nt) / fs
    f0 = 20.
    signal = chirp(t, f0, t[-1], 100.0)
    #signal *= (1.0 + 0.5 * np.sin(2.0*np.pi*3.0*t))
    fsignal =  np.linspace(20,100,nt)
    
    fmax = np.max(fsignal)
    fmin = np.min(fsignal)
    
# Case 1: self coded chirp (with same parameters as in case 0)
if icase == 1:
    duration = 1.0
    fs = 400.0
    nt = int(fs*duration)
    t = np.arange(nt) / fs
    f0 = 20.
    f1 = 100.
    t1 = t[-1]
    
    fm = f0 + (f1 - f0) * t / t1
    
    ifm = np.zeros(nt)
    for i in range(nt):
        ifm[i] = 2*np.pi*np.sum(fm[:i])
        
    ifm = ifm / fs

    signal = np.cos(ifm)
       
    fsignal = fm
    
    fmax = np.max(fm)
    fmin = np.min(fm)

# Case 2: self coded chirp (with ring laser  parameters , linear progression)
if icase == 2:
    duration = 2.0
    fs = 5000.0
    nt = int(fs*duration)
    t = np.arange(nt) / fs
    f0 = 340.
    f1 = 360.
    t1 = t[-1]
    
    # ORiginal
    fm = f0 + (f1 - f0) * t / t1
    
    ifm = np.zeros(nt)
    for i in range(nt):
        ifm[i] = 2*np.pi*np.sum(fm[:i])
        
    ifm = ifm / fs

    signal = np.cos(ifm)
    fsignal = fm

        
    fmax = np.max(fm)
    fmin = np.min(fm)

# Case 3: self coded chirp (with ring laser  parameters , sinusoidal variations)
if icase == 3:
    duration = 2.0
    fs = 5000.0
    nt = int(fs*duration)
    t = np.arange(nt) / fs
    f0 = 350.

    fm = f0 + np.sin(2*np.pi*1*t) 
    
    ifm = np.zeros(nt)
    for i in range(nt):
        ifm[i] = 2*np.pi*np.sum(fm[:i])
        
    ifm = ifm / fs

    signal = np.cos(ifm)
    fsignal = fm
    
    
    fmax = np.max(fm-f0)
    fmin = np.min(fm-f0)
    
# Case 4: self coded  chirp (with ring laser  parameters , random variation)
if icase == 4:
    duration = 2.0
    fs = 5000.0
    nt = int(fs*duration)
    t = np.arange(nt) / fs
    f0 = 350.
    
    perturbation = np.sin(2*np.pi*3*t) + np.sin(2*np.pi*1*t) + np.sin(2*np.pi*5*t) + np.sin(2*np.pi*10*t)

    fm = f0 + perturbation
    
    ifm = np.zeros(nt)
    for i in range(nt):
        ifm[i] = 2*np.pi*np.sum(fm[:i])
        
    ifm = ifm / fs

    signal = np.cos(ifm)
    fsignal = fm

     # min max
    fmax = np.max(fm-f0)
    fmin = np.min(fm-f0)
    
# Case 5: with rotation rate signal from G ring
if icase == 5:

    fs = 5000.    # sampling rate final Sagnac signal
    
    # Desired time stamping
    t = np.linspace(0, torig.max(), int(torig.max()*fs))
    nt = np.size(t)
    
    # Interpolate real data
    print(' Interpolating ')
    rdata = np.interp(t,torig,data)  


# Extract window 

    # artificial amplification
    
    amp = 1000
    f0 = fSagnac

    #Instantaneous frequency to recover
    fm = f0 +  4*A/(l*P)*rdata * amp
    
    ifm = np.zeros(nt)
    
    print(' Integrating ... ')
    for i in range(nt):
        ifm[i] = 2*np.pi*np.sum(fm[:i])
        
    ifm = ifm / fs

    signal = np.cos(ifm)
    fsignal = fm
    
    # min max
    fmax = np.max(fm-f0)
    fmin = np.min(fm-f0)
    
 
```

## Frequency Demodulation

```python
## Frequency demodulation (ignoring what is happening inside), 
## taken from tutorial

analytic_signal = hilbert(signal)

amplitude_envelope = np.abs(analytic_signal)

instantaneous_phase = np.unwrap(np.angle(analytic_signal))

instantaneous_frequency = (np.diff(instantaneous_phase) * (2.0*np.pi) * fs)
```

```python
## Plotting
plt.rcParams['figure.figsize'] = 8, 6
fig = plt.figure()
ax0 = fig.add_subplot(211)
ax0.plot(t, signal, label='signal')
ax0.plot(t, amplitude_envelope, label='envelope')
ax0.set_xlabel("time in seconds")
ax0.set_xlim(0, 100*1/fs)
ax0.legend()
ax0.set_title('Signal and envelope')
ax1 = fig.add_subplot(212)
if icase > 2:
    ax1.plot(t[1:], instantaneous_frequency-f0,label = ' demod. Signal ')
else:
    ax1.plot(t[1:], instantaneous_frequency,label = ' demod. Signal ')
if icase > 2:
    ax1.plot(t, fsignal-f0,label = ' demod. Signal ')
else:
    ax1.plot(t, fsignal,label = ' demod. Signal ')
ax1.legend()
ax1.set_xlabel("time in seconds")
ax1.set_ylim(fmin, fmax)
ax1.set_title('Instantaneous Frequency')
plt.tight_layout() 

# put the plot size back to normal
plt.rcParams['figure.figsize'] = 8, 3
```
