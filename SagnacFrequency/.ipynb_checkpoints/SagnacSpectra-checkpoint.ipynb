```python
def __make_decibel(array, relative_value):
    return 20*log10(array/relative_value)
```

## Configurations

```python
date = UTCDateTime("2021-03-05")

config = {}

config['seed_id'] = "BW.DROMY..FJZ"

# psd = zeros([])

config['opath'] = "/home/andbro/Documents/ROMY/SagnacSpectra/plots/"

config['repository'] = "george"

config['ring'] = "Z"

config['rings'] = {"Z":553, "U":302, "V":448,"W":448}

config['f_expected'] = config['rings'][config['ring']]  ## expected sagnac frequency
config['f_band'] = 10 ## +- frequency band

config['interval']= 3600

config['tbeg'] = date
config['tend'] = date + config['interval']

config['f_lower'], config['f_upper'] = config['f_expected']-100, config['f_expected']+100

config['segment_factor'] = 1e4
```

## Looping 

```python
def __get_welch_psd(config, arr, df):

    segment_factor = df*1200

    f0, psd0 = welch(  
            arr, 
            fs=df, 
            window='hanning', 
            nperseg=segment_factor, 
            noverlap=int(segment_factor/2), 
            nfft=None, 
            detrend='constant', 
            return_onesided=True, 
            scaling='density',
            )

    return f0, psd0
```

```python
def __looping(config):

    from numpy import where
    
    psd = []
    tbeg0 = config['tbeg']
    tend0 = tbeg0+config['interval']
    
    while tbeg0.hour < 4 and tbeg0.date == date:
        
        print(f"{tbeg0.hour} - {tbeg0.hour+1}")

        st, inv = __querrySeismoData(    
                                    seed_id=config['seed_id'],
                                    starttime=tbeg0-60,
                                    endtime=tend0+60,
                                    repository=config['repository'],
                                    path=None,
                                    restitute=False,
                                    detail=None,
                                    )
        
        st.trim(tbeg0, tend0)

        st[0].data = st[0].data * 0.59604645e-6

        f, psd0 = __get_welch_psd(config, st[0].data, st[0].stats.sampling_rate)     
        
#         f, psd0 = welch(  
#                         st[0].data, 
#                         fs=st[0].stats.sampling_rate, 
#                         window='hanning', 
#                         nperseg=256*config['segment_factor'], 
#                         noverlap=128*config['segment_factor'], 
#                         nfft=None, 
#                         detrend='constant', 
#                         return_onesided=True, 
#                         scaling='density',
#                        )

        idx1 = where(f <= config['f_lower'])[0][0]
        idx2 = where(f >= config['f_upper'])[0][0]

        psd.append(psd0[idx1:idx2])
        f = f[idx1:idx2]   

#         psd.append(psd0)
        del st

        tbeg0 += config['interval']
        tend0 += config['interval']


    psd = array(psd)
    
    return f, psd

f, psd = __looping(config)
```

## Colorline Plot

```python
def __makeplot_colorlines(config, ff, data):

    
    def __get_median_psd(psds):

        from numpy import median, zeros, isnan

        med_psd = zeros(psds.shape[1])

        for f in range(psds.shape[1]):
            a = psds[:,f]
            med_psd[f] = median(a[~isnan(a)])

        return med_psd

    cols = plt.cm.jet_r(linspace(0,1,shape(data)[0]+1))
#     cols = plt.cm.viridis(linspace(0,1,shape(data)[0]+1))
    
    ## ____________________________________________

    fig, ax = plt.subplots(1,1, figsize=(15,10))

    font = 14
    
    data_min = min([min(d) for d in data])
    data_max = max([max(d) for d in data])

    for i, psdx in enumerate(data):

        ax.plot(ff, psdx, color=cols[i], zorder=2, label=i,  alpha=0.3)
#         ax.scatter(ff, psdx, color=cols[i], s=1,  alpha=0.7, label=i,)

    ax.plot(ff, __get_median_psd(data), color='k', lw=1, zorder=2)

    ax.set_xlim(config['f_expected']-config['f_band'], config['f_expected']+config['f_band'])

    ax.set_yscale("log")
    ax.set_ylim(data_min-0.01*data_min, data_max+0.01*data_max)
    
    ax.legend(ncol=2)

    ax.grid(ls='--', zorder=1)

    ax.set_xlabel("Frequency (Hz)", fontsize=font)
    ax.set_ylabel(f"PSD (V$^2$ /Hz) ", fontsize=font)
    ax.set_title(f"Sagnac Spetra on {date.date} ({config['interval']}s windows) ", fontsize=font+2)

    ax.tick_params(axis='both', labelsize=font-2)

    plt.show();
    return fig

fig = __makeplot_colorlines(config, f, psd);
```

```python
fig.savefig(
            config['opath']+f"plot_sagnacspectra_{date.date}_{config['interval']}_colorlines.png", 
            dpi=300, facecolor='w', edgecolor='w', orientation='portrait', 
            format="png", transparent=False, bbox_inches="tight", pad_inches=0.2,
           )
```
