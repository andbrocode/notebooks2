```python
def __get_welch_psd(config, arr, df):

    segments = df*config['segment_factor']

    f0, psd0 = welch( 
                    arr, 
                    fs=df, 
                    window='hanning', 
                    nperseg=segments, 
                    noverlap=int(segments/2), 
                    nfft=None, 
                    detrend='constant', 
                    return_onesided=True, 
                    scaling='density',
                    )

    return f0, psd0
```

## Looping 

```python
def __looping(config):

    from numpy import where
    
    psd = []
    tbeg0 = config['tbeg']
    tend0 = tbeg0+config['interval']
    
    while tbeg0.hour < 6 and tbeg0.date == date:
        
        print(f"{tbeg0.hour} - {tbeg0.hour+1}")

        st, inv = __querrySeismoData(    
                                    seed_id=config['seed_id'],
                                    starttime=tbeg0-60,
                                    endtime=tend0+60,
                                    repository=config['repository'],
                                    path=None,
                                    restitute=False,
                                    detail=None,
                                    )
        
        st.trim(tbeg0, tend0)

        st[0].data = st[0].data * 0.59604645e-6

        f, psd0 = __get_welch_psd(config, st[0].data, st[0].stats.sampling_rate)     
        
#         f, psd0 = welch(  
#                         st[0].data, 
#                         fs=st[0].stats.sampling_rate, 
#                         window='hanning', 
#                         nperseg=256*config['segment_factor'], 
#                         noverlap=128*config['segment_factor'], 
#                         nfft=None, 
#                         detrend='constant', 
#                         return_onesided=True, 
#                         scaling='density',
#                        )

        idx1 = where(f <= config['f_lower'])[0][0]
        idx2 = where(f >= config['f_upper'])[0][0]

        psd.append(psd0[idx1:idx2])
        f = f[idx1:idx2]   

#         psd.append(psd0)
        del st

        tbeg0 += config['interval']
        tend0 += config['interval']


    psd = array(psd)
    
    return f, psd

f, psd = __looping(config)
```

```python
def __makeplot_colorlines(config, ff, data, peaks=None, smooth=None):

    from numpy import log10, median
    
    def __get_median_psd(psds):

        from numpy import median, zeros, isnan

        med_psd = zeros(psds.shape[1])

        for f in range(psds.shape[1]):
            a = psds[:,f]
            med_psd[f] = median(a[~isnan(a)])

        return med_psd

    def __get_peaks(arr, smoothing=10):
    
        from scipy.signal import find_peaks, peak_widths
        from numpy import argmax, log10

        def __smooth(y, box_pts):
            from numpy import ones, convolve, hanning

            win = hanning(box_pts)
            y_smooth = convolve(y, win/sum(win), mode='same')

            return y_smooth

        data = __smooth(log10(psd[0]), smoothing)

#         peaks, _ = find_peaks(data, prominence=1.4, distance=50)
        peaks, _ = find_peaks(data, threshold=1, distance=20)

        return peaks
    
    
    cols = plt.cm.jet_r(linspace(0,1,shape(data)[0]+1))
#     cols = plt.cm.viridis(linspace(0,1,shape(data)[0]+1))
    
    ## ____________________________________________

    fig, ax = plt.subplots(1,1, figsize=(15,10))

    font = 14
    
    data_min = min([min(d) for d in data])
    data_max = max([max(d) for d in data])

    for i, psdx in enumerate(data):
        
        if smooth is not None:
            ax.plot(ff, __smooth(psdx,smooth), color=cols[i], zorder=2, label=i,  alpha=0.3)
        else:
            ax.plot(ff, psdx, color=cols[i], zorder=2, label=i,  alpha=0.3)

    ## select only psds above a median threshold for median computation
    psd_select = array([dat for dat in data if median(log10(dat)) > -8])
    psd_median = __get_median_psd(psd_select)
    ax.plot(ff, psd_median, color='k', lw=1, zorder=2)

    
    if peaks:
        pp = __get_peaks(psd_median, smoothing=smooth)
        for p in pp:
            if ff[p] < 553.37:
                ax.annotate(f"{round(ff[p],2)}", (ff[p], psd_median[p]), (ff[p]-1.5, psd_median[p]+0.1*psd_median[p]), 
                            bbox = dict(boxstyle ="round", fc ="0.9"), 
                            arrowprops = dict(arrowstyle = "->", connectionstyle = "angle, angleA = 0, angleB = 90,rad = 10")
                           )
            else:
                ax.annotate(f"{round(ff[p],2)}", (ff[p], psd_median[p]), (ff[p]+0.5, psd_median[p]+0.1*psd_median[p]), 
                            bbox = dict(boxstyle ="round", fc ="0.9"), 
                            arrowprops = dict(arrowstyle = "->", connectionstyle = "angle, angleA = 0, angleB = 90,rad = 10")
                           )                
    
    ax.set_xlim(config['f_expected']-config['f_band'], config['f_expected']+config['f_band'])

    ax.set_yscale("log")
    ax.set_ylim(data_min-0.01*data_min, data_max+0.5*data_max)
    
    leg = ax.legend(ncol=2)

    # change the line width for the legend
    [ line.set_linewidth(3.0) for line in leg.get_lines()]


    ax.grid(ls='--', zorder=1)

    ax.set_xlabel("Frequency (Hz)", fontsize=font)
    ax.set_ylabel(f"PSD (V$^2$ /Hz) ", fontsize=font)
    ax.set_title(f"Sagnac Spetra on {date.date} ({config['interval']}s windows) ", fontsize=font+2)

    ax.tick_params(axis='both', labelsize=font-2)

    plt.show();
    return fig

fig = __makeplot_colorlines(config, f, psd, peaks=True, smooth=20);
```

```python
def __makeplot_distribution(config, xx, yy, dist, overlay=False):

    from numpy import nanmax, nanmin
    from matplotlib import colors

    def __smooth(y, box_pts):
        from numpy import ones, convolve, hanning

        win = hanning(box_pts)
        y_smooth = convolve(y, win/sum(win), mode='same')

        return y_smooth


    
    cmap = plt.cm.get_cmap("YlOrRd")
    cmap = plt.cm.get_cmap("viridis_r")
    cmap.set_bad("white")
    cmap.set_under("white")

    max_psds = nanmax(dist)
    min_psds = nanmin(dist)

   
    ## ____________________________________________

    fig, ax = plt.subplots(1,1, figsize=(15,10))

    plt.subplots_adjust(wspace=0.1)
    
    font = 14

    im = ax.pcolormesh( xx, yy, dist.T, 
                        cmap=cmap,
                        vmax=max_psds,
                        vmin=min_psds+0.01*min_psds,
#                         norm=colors.LogNorm(),
                        )     

    if overlay is not None:
        ax.plot(xx, __smooth(10**overlay, 50), color='k', alpha=0.6, lw=1, zorder=2, label="maxima")

    ax.set_xlim(config['f_expected']-config['f_band'], config['f_expected']+config['f_band'])

    ax.set_yscale("log")
    
    ax.legend(ncol=2)

    ax.grid(ls='--', zorder=1)

    ax.set_xlabel("Frequency (Hz)", fontsize=font)
    ax.set_ylabel(f"PSD (V$^2$ /Hz) ", fontsize=font)
    ax.set_title(f"Sagnac Spetra on {date.date} ({config['interval']}s windows) ", fontsize=font+2)

    ax.tick_params(axis='both', labelsize=font-2)

    cb = plt.colorbar(im, ax=ax, anchor=(-0.5, 0.05))
    cb.set_label("Propability Density", fontsize=font, labelpad=-50)
    
    plt.show();
    return fig

fig = __makeplot_distribution(config, f, out['bin_mids'], out['dist'], overlay=out['bins_maximas']);
```
