## Configurations

```python
# specify path to psds median data
path_to_data = archive_path+"ModalAnalysis/data/PSDS_median/"

# specify path to store output figures
path_to_figs = archive_path+f"ModalAnalysis/figures/"
```

## Load and Process Data

```python
psds = {}

# "RY03" is distorted
# "RY04" missing data
stas = ["FFB1", "FFB2", "FFB3",
        "RY01", "RY02", "RY05", "RY06", "RY07", "RY08", "RY09"]

for i, sta in enumerate(stas):

    print(f"{sta} ...")
    try:
        _psd = read_pickle(path_to_data+f"{sta}_psd_stats.pkl")
    except:
        print(f" -> failed for {sta}")
        continue

    if i == 0:
        psds['ff'] = _psd.frequencies

    psds[sta] = {}

    for c in ["Z", "N", "E"]:
        try:
            psds[sta][c] = _psd[f"psds_median_{c}"].values
        except:
            print(f" -> all NaN for {sta}{c}")
            psds[sta][c] = psds['ff']*np.nan
```

## Plotting

```python
# get colors
colors = plt.cm.tab20(range(len(stas)))
```

```python
def __makeplot():

    Nrow, Ncol = 3, 1

    font = 12

    fig, ax = plt.subplots(Nrow, Ncol, figsize=(15, 8), sharex=True)

    plt.subplots_adjust(hspace=0.1)

#     for sta in stas:
#         ax[0].loglog(psds["ff"], psds[sta]["Z"], label=f"{sta}")

#     for sta in stas:
#         ax[1].loglog(psds["ff"], psds[sta]["N"], label=f"{sta}")

#     for sta in stas:
#         ax[2].loglog(psds["ff"], psds[sta]["E"], label=f"{sta}")

    for i, sta in enumerate(stas):
        for k, c in enumerate(["Z", "N", "E"]):
            try:
                ax[k].loglog(psds["ff"],
                             psds[sta][c],
                             label=f"{sta}",
                             color=colors[i]
                            )
            except:
                print(f" -> skip {sta}")
                continue

    for _n in range(Nrow):
        ax[_n].grid(which="both", ls=":", zorder=0)
        ax[_n].set_yscale('log')
        ax[_n].set_xscale('log')
        ax[_n].set_xlim(5e-1, 1e2)
        ax[_n].set_ylim(1e-19, 1e-13)
        ax[_n].set_ylabel(r"PSD (m$^2$/s$^2$/Hz)", fontsize=font)

    ax[2].set_xlabel("Frequency (Hz)", fontsize=font)

    ax[0].legend(bbox_to_anchor=(0.5, 1.1), loc="center", ncol=len(stas)+len(stas))

    ## panel labels
    ax[0].text(-.08, 1.05, '(a)', ha='left', va='top', transform=ax[0].transAxes, fontsize=font+2)
    ax[1].text(-.08, 1.05, '(b)', ha='left', va='top', transform=ax[1].transAxes, fontsize=font+2)
    ax[2].text(-.08, 1.05, '(c)', ha='left', va='top', transform=ax[2].transAxes, fontsize=font+2)

    ax[0].text(.96, .97, 'HHZ', ha='left', va='top', transform=ax[0].transAxes, fontsize=font)
    ax[1].text(.96, .97, 'HHN', ha='left', va='top', transform=ax[1].transAxes, fontsize=font)
    ax[2].text(.96, .97, 'HHE', ha='left', va='top', transform=ax[2].transAxes, fontsize=font)

    plt.show();
    return fig
```

```python
fig = __makeplot();

fig.savefig(path_to_figs+f"RY_spectral_averages.png", format="png", dpi=150, bbox_inches='tight')
```

### Smoothing

```python
psds_smth = {}

NN = 150

psds_smth['ff'] = __smooth(psds['ff'], NN)

for sta in stas:
    if sta == "ff":
        continue
    psds_smth[sta] = {}
    for c in ["Z", "N", "E"]:
        try:
            psds_smth[sta][c] = __smooth(psds[sta][c], NN)
        except:
            continue
```

```python
def __makeplot2():

    Nrow, Ncol = 3, 1

    font = 12

    fig, ax = plt.subplots(Nrow, Ncol, figsize=(15, 8), sharex=True)

    plt.subplots_adjust(hspace=0.1)

    for i, sta in enumerate(stas):
        for k, c in enumerate(["Z", "N", "E"]):
            try:
                ax[k].plot(psds_smth["ff"],
                           psds_smth[sta][c],
                           label=f"{sta}",
                           color=colors[i],
                           )
            except:
                continue

    for _n in range(Nrow):
        ax[_n].grid(which="both", ls=":", zorder=0)
        ax[_n].set_yscale('log')
        ax[_n].set_xscale('log')
        ax[_n].set_xlim(5e-1, 1e2)
        ax[_n].set_ylim(1e-19, 1e-13)
        ax[_n].set_ylabel(r"PSD (m$^2$/s$^2$/Hz)", fontsize=font)

    ax[2].set_xlabel("Frequency (Hz)", fontsize=font)

    ax[0].legend(bbox_to_anchor=(0.5, 1.1), loc="center", ncol=len(stas))

    # panel labels
    ax[0].text(-.08, 1.05, '(a)', ha='left', va='top', transform=ax[0].transAxes, fontsize=font+2)
    ax[1].text(-.08, 1.05, '(b)', ha='left', va='top', transform=ax[1].transAxes, fontsize=font+2)
    ax[2].text(-.08, 1.05, '(c)', ha='left', va='top', transform=ax[2].transAxes, fontsize=font+2)

    ax[0].text(.96, .97, 'HHZ', ha='left', va='top', transform=ax[0].transAxes, fontsize=font)
    ax[1].text(.96, .97, 'HHN', ha='left', va='top', transform=ax[1].transAxes, fontsize=font)
    ax[2].text(.96, .97, 'HHE', ha='left', va='top', transform=ax[2].transAxes, fontsize=font)

    plt.show();
    return fig
```

```python
fig = __makeplot2();

fig.savefig(path_to_figs+f"RY_spectral_averages_smoothed.png", format="png", dpi=150, bbox_inches='tight')
```

### H over V

```python
%matplotlib inline
```

```python
peaks = [0.228, 1.13, 1.24, 2.085, 2.775, 3.125, 16.66]
```

```python
def __makeplot3(smooth=True):

    Nrow, Ncol = 1, 1

    font = 12

    fig, ax = plt.subplots(Nrow, Ncol, figsize=(15, 8), sharex=True)

    plt.subplots_adjust(hspace=0.1)

    for i, sta in enumerate(stas):

        try:
            ff = psds_smth["ff"]

            psdsZ = psds[sta]["Z"]
            psdsN = psds[sta]["N"]
            psdsE = psds[sta]["E"]

            psdsH = np.sqrt(psdsN**2+psdsE**2)

            if smooth:
                HoverV = __smooth(psdsH/psdsZ, 150)
            else:
                HoverV = psdsH/psdsZ
            
            ax.plot(ff, HoverV, label=f"{sta}", color=colors[i])
            
        except:
            continue

    for peak in peaks:
        y1, y2 = ax.get_ylim()
        ax.axvline(peak, y1, y2, color="yellow", zorder=0)

    ax.grid(which="both", ls=":", zorder=0)
    ax.set_yscale('log')
    ax.set_xscale('log')
    ax.set_xlim(2e-1, 5e1)
    ax.set_ylim(3e-2, 1.2e1)
    ax.set_ylabel(r"H / V  = $\sqrt{N^2 + E^2}$ / Z ", fontsize=font)

    ax.set_xlabel("Frequency (Hz)", fontsize=font)

    ax.legend(bbox_to_anchor=(0.5, 1.04), loc="center", ncol=len(stas))

    plt.show();
    return fig
```

```python
fig = __makeplot3(smooth=True);

fig.savefig(path_to_figs+f"RY_spectral_HoverV_smoothed.png", format="png", dpi=150, bbox_inches='tight')
```
